```python
import pandas as pd
from google.colab import drive

# ✅ توصيل Google Drive
drive.mount('/content/drive')
# ✅ تحميل البيانات
df = pd.read_csv("/content/drive/My Drive/FE462_final_file.csv", parse_dates=['REGN_DATE'])

# ✅ ترتيب البيانات لضمان الدمج الصحيح
df = df.sort_values(by=['SAMPLE_ID', 'REGN_DATE'])

# ✅ دمج الصفوف بناءً على القواعد المحددة
merged_rows = []
skip_indexes = set()

for i in range(len(df)):
    if i in skip_indexes:  # تخطي الصفوف التي تم دمجها بالفعل
        continue

    row = df.iloc[i].copy()
    sample_id = row['SAMPLE_ID']
    regn_date = row['REGN_DATE']

    # البحث عن الصفوف المطابقة للأعمدة المحددة، للرنج بيوم أو يومين
    similar_rows = df[(df['SAMPLE_ID'] == sample_id) &
                      (df['REGN_DATE'].between(regn_date - pd.Timedelta(days=2), regn_date + pd.Timedelta(days=2)))]

    if len(similar_rows) > 1:
        for col in df.columns:
            if col not in ['SAMPLE_ID', 'REGN_DATE']:  # عدم تعديل الأعمدة الأساسية
                row[col] = similar_rows[col].dropna().iloc[0] if similar_rows[col].dropna().shape[0] > 0 else None

        # إضافة الفهارس إلى قائمة التجاهل لتجنب تكرار الدمج
        skip_indexes.update(similar_rows.index)

    merged_rows.append(row)

# ✅ إنشاء DataFrame جديد بالبيانات المدمجة
df_merged = pd.DataFrame(merged_rows)

# ✅ حفظ الملف بعد التنظيف
df_merged.to_csv("path_to_save_hello_file.csv", index=False)

# ✅ طباعة عدد الصفوف والأعمدة بعد الدمج
print(f"📊 عدد الصفوف بعد الدمج: {df_merged.shape[0]}")
print(f"📊 عدد الأعمدة: {df_merged.shape[1]}")
```

```python
import pandas as pd
from google.colab import drive

# 📌 1️⃣ تحميل Google Drive
drive.mount('/content/drive')

# 📌 2️⃣ تحديد مسار الملف
file_path = '/content/drive/My Drive/FE462.csv'

# 📌 3️⃣ تحميل البيانات
df = pd.read_csv(file_path)
print(f"قبل التنظيف: {df.shape[0]} صف، {df.shape[1]} عمود")

df = df.drop(df.columns[0], axis=1)
# 📌 5️⃣ حذف الصفوف المكررة بالكامل
df_cleaned = df.drop_duplicates()

# 📌 6️⃣ طباعة عدد الصفوف قبل وبعد التنظيف
print(f"🔹 عدد الصفوف قبل التنظيف: {len(df)}")
print(f"🔹 عدد الصفوف بعد التنظيف: {len(df_cleaned)}")

# 📌 7️⃣ دمج الصفوف بناءً على تطابق الأعمدة 'SAMPLE_ID' و 'AGE_MONTHS'
df_merged = df_cleaned.groupby(['SAMPLE_ID', 'AGE_MONTHS'], as_index=False).agg(
    lambda x: x.dropna().iloc[0] if x.dropna().shape[0] > 0 else None
)

# 📌 8️⃣ حفظ الملف المدمج في Google Drive
output_path = "/content/drive/My Drive/FE462_Merged.csv"
df_merged.to_csv(output_path, index=False)

# 📌 9️⃣ طباعة عدد الصفوف والأعمدة بعد الدمج
print(f"✅ عدد الصفوف بعد الدمج: {df_merged.shape[0]}")
print(f"✅ عدد الأعمدة: {df_merged.shape[1]}")

# 📌 🔟 عرض أول 5 صفوف من الملف النهائي
df_merged.head()
```
